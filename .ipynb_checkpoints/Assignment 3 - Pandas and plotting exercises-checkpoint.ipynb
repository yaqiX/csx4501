```python
#I'm interested in the economic status in 2008 when the financial crisis happened
crisis = df[(df['Year'] == 2008)&['State']]
print(crisis)
```

```python
#I'm interested in the Data.Calculated.Net Job Creation under the crisis
crisis.plot(x='State',y='Data.Calculated.Net Job Creation',kind='bar',figsize=(50,10),fontsize=30)
#it shows that there were very little states have good net job creation under the crisis.
```

```python
# Somehow Washington state had exceptionally high net job creation, so I'm curious why.
# I think I can check the trend of Net Job Creation of Washington statedf.
wash = df[(df['State'] == 'Washington')]
wash.plot(x='Year',y='Data.Calculated.Net Job Creation',kind='barh',figsize=(10,20),fontsize=10)
#it turned out there was a sharp decrese after 2008 in Washington, it was just one year late.  
```

```python
# Hawaii looked sad in 2008. I'm also curious what happened.
haw = df[(df['State'] == 'Hawaii')]
haw.plot(x='Year',y='Data.Calculated.Net Job Creation',kind='barh',figsize=(10,20),fontsize=10)
#It seems that Hawaii had not been good for years. I'm curious if this affect company establishment. As the saying goes,"Crisis brings opportunity"
```

```python
haw.plot(x='Year',y='Data.Establishments.Entered',kind='line',figsize=(20,10),fontsize=30)
# There was a sharp growth of company establishment before the crisis, and actually after the crisis, the number of new established company never had a "boom".
# I wonder if a lot of company died because of the crisis
```

```python
# I wonder if a lot of company died because of the crisis
haw.plot(x='Year',y='Data.Establishments.Exited',kind='line',figsize=(20,10),fontsize=30)
# The result shows yes.
```
